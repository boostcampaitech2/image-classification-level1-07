```python
import re
import pandas as pd

# 용범님이 주신 validation에 맞게 csv만들기
trainyb = pd.read_csv('/opt/ml/input/data/train/train_yb.csv')
trainyb = trainyb[['img_path', 'person_label']]
trainyb = trainyb.rename(columns = {'img_path': 'path', 'person_label': 'class'})

trainyb['path'] = trainyb['path'].map(lambda x : re.sub(r'images', 'face_crop',x))

utk = pd.read_csv('/opt/ml/newinput/UTKFace_only.csv')
tot = pd.concat([trainyb, utk])
print(len(tot))
tot.head()
```
